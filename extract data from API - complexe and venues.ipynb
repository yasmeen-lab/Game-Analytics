```python
import requests
import csv

api_key = "33HrNbPTOHP9LF57RNd4d8yHYBfKXVZZ0XFAgxFq"
url = f"https://api.sportradar.com/tennis/trial/v3/en/complexes.json?api_key={api_key}"
headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    data = response.json()
    complexes = data.get("complexes", [])

    with open("venues_table.csv", "w", newline='', encoding='utf-8') as f:
        writer = csv.writer(f)

        # ✅ Match CSV columns with your SQL table
        writer.writerow([
            "venue_id", "venue_name", "city_name", "country_name", 
            "country_code", "timezone", "complex_id"
        ])

        for comp in complexes:
            complex_id = comp.get("id", "")
            venues = comp.get("venues", [])

            for venue in venues:
                writer.writerow([
                    venue.get("id", ""),
                    venue.get("name", ""),
                    venue.get("city", ""),
                    venue.get("country", {}).get("name", ""),
                    venue.get("country_code", ""),
                    venue.get("timezone", ""),
                    complex_id
                ])

    print("✅ Venues data saved to venues_table.csv")

else:
    print(f"Failed to fetch data: {response.status_code}\n{response.text}")
```
